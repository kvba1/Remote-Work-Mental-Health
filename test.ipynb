```python
# Import bibliotek
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Wczytaj dane
data = pd.read_csv('Impact_of_Remote_Work_on_Mental_Health.csv')

# 1. Podstawowe informacje o danych
print("Podstawowe informacje o danych:")
display(data.info())

# 2. Sprawdzenie brakujących danych
missing_data = data.isnull().sum()
print("\nBrakujące dane w każdej kolumnie:")
display(missing_data)
# Tworzymy DataFrame z brakującymi danymi
missing_data_df = pd.DataFrame(missing_data, columns=['Missing Values']).reset_index()
missing_data_df.columns = ['Column', 'Missing Values']

# Wyświetlamy tabelę brakujących danych
display(missing_data_df)


# 3. Podstawowe statystyki dla danych liczbowych i kategorycznych
print("\nPodstawowe statystyki dla danych liczbowych:")
display(data.describe())

print("\nPodstawowe statystyki dla danych kategorycznych:")
display(data.describe(include=['object']))

# 4. Wizualizacja brakujących danych (jeśli istnieją)
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Wizualizacja brakujących danych")
plt.show()

# 8. Analiza korelacji tylko dla kolumn liczbowych
numeric_data = data.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title("Macierz korelacji dla danych liczbowych")
plt.show()
# Obliczenie macierzy korelacji dla kolumn liczbowych
correlation_matrix = numeric_data.corr()

# Znalezienie maksymalnych i minimalnych wartości korelacji, wyłączając autokorelacje (diagonalne)
max_correlation = correlation_matrix[correlation_matrix != 1].max().max()
min_correlation = correlation_matrix[correlation_matrix != 1].min().min()

print("Maksymalna wartość korelacji:", max_correlation)
print("Minimalna wartość korelacji:", min_correlation)


```
